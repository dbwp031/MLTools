```python
import csv
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('voc_all.csv', encoding='UTF8')
```

```python
g = sns.scatterplot(data=data, x= 'FLOPs (G)', y = 'mean IoU (%)')
g.axhline(y=70.3, color='r', linewidth=1,linestyle='--',label='1% drop')
g.axhline(y=66.3, color='b', linewidth=1,linestyle='--',label='5% drop')

plt.legend()
plt.savefig('voc_all.png',dpi=200)
plt.show()
```
